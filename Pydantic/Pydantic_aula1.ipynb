Importação dos modulos, bem como suas ferramentas que serão trabalhadas.

```python
from enum import auto, IntFlag
from typing import Any

from pydantic import (
    BaseModel,
    EmailStr,
    Field,
    SecretStr,
    ValidationError
)
```

Criação da classe Role (Função), que vem a ser herdada do IntFlag para auxiliar na definição do que o usuário pode ser. Além disso, temos a criação da classe User, que serve para determinar como deve ser a estrutura dos dados do usuário. Nessa clase, temos a utilização do Field para ajudar na "configuração" dos atributos, adicionando regras e limites. No caso do "name", O Field estabelece um exemplo de nome para auxiliar na documentação. Já no caso do "email", o Field além de estabelecer um exemplo, também proporciona uma descrição que pode ser usado em uma chamada de API e o parametro "frozen=True" que torna imutavel a estrutura do atributo. Por fim, temos a configuração do "password" e o "role" do usuário que foi criado anteriormente.

```python
!pip install 'pydantic[email]'

class Role(IntFlag):
    Author = auto()
    Editor = auto()
    Developer = auto()
    Admin = Author | Editor | Developer

class User(BaseModel):
  name: str = Field(examples=["Arjan"])
  email: EmailStr = Field(
        examples=["example@rjancodes.com"],
        description="The email address of the user",
        frozen=True
        )
  password: SecretStr = Field(
      examples=["Passaword123"], description="The password of the user"
  )
  role: Role = Field(default=None, description="The role of the user")
```

Nessa parte, temos a criação da função validate que irá fazer a validação dos dados. Caso os dados estiverem certos, a função exibe esses dados, o que significa que os dados passados estão de acordo com a estrura da classe User. No entanto, se os dados estiverem errados, a função exibe uma mensagem dizendo "User is invalid" e indica os erros na validação. Por fim, temos a criação da função main, que serve apenas para controlar toda a operação, chamando a função validade e passando um exemplo de dados bom e um exemplo de dados ruim.

```python
def validate(data: dict[str, Any]) -> None:
  try:
    user = User.model_validate(data)
    print(user)
  except ValidationError as e:
    print("User is invalid")
    for error in e.errors():
      print(error)

def main() -> None:
  good_data = {
      "name": "Arjan",
      "email": "example@rjancodes.com",
      "password": "Passaword123",
  }
  bad_data = {"email": "<bad data>", "password": "<bad_data>"}

  validate(good_data)
  validate(bad_data)

if __name__ == "__main__":
  main()
```
