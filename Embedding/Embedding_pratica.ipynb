Importação dos módulos necessário para implementação de uma busca semântica em um banco de dados vetorial.

```python
from qdrant_client import QdrantClient
from qdrant_client.models import Distance, VectorParams
from qdrant_client.models import PointStruct
from fastembed import TextEmbedding
```

Estabelece a conexão com o Qdrant, que é um banco de dados vetorial - armazena palavras/informações em forma de vetores númericos. Também foi feita a criação de uma coleçaõ de no Qdrant, determinando que o armazenado dos vetores deve ser de 384 dimensões, além de configurar a distância do cosseno para o parametro de similaridade.

```python
client = QdrantClient(url="sua_url", api_key="sua_api_key")

client.create_collection(collection_name="linguagem", vectors_config=VectorParams(size=384, distance=Distance.COSINE))
```

Por fim, temos o estabelecimento do que queremos buscar (query_text) no banco de dados vetorial que criamos anteriormente com o Qdrant, e uma relação de quais dessas informações no menu mais se assemelham com aquilo que estamos buscando.

```python
query_text = "RAG"
query_vector = next(iter(model.embed(query_text)))

results = client.query_points(
    collection_name="linguagem",
    query=query_vector,
    with_payload=True,
    limit=5
)

for result in results.points:
    print(f"Titiulo: {result.payload.get('titulo', 'N/A')}")
    print(f"Score: {result.score}")
    print(f"Descricao: {result.payload['descricao'][:150]}...")
    print(f"Tipo: {result.payload.get('tipo', 'N/A')}")
    print("---")
```
